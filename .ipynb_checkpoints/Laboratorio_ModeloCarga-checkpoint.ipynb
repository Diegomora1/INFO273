# Notebook 1: Modelo de Carga
En este notebook vamos a implementar el algoritmo de clustering jerárquico en R con ayuda de la función hclust. Utilizaremos como ejemplo el modelo de carga web visto durante la sesión y cuyos datos se encuentran en el archivo cargaWeb.csv

```r
#ruta datos
#lee datos del archivo de texto como una matriz. La primera #fila se utiliza 
#como vector de nombres de las columnas
datos <-read.table("cargaWeb.csv",dec=".",sep=";",header=TRUE)
print(datos)
summary(datos)
```

```r
d <- dist(datos)
hd <- hclust(d, method = "single", members=NULL)
plot(hd, labels = NULL, hang = 0.1, axes = TRUE, frame.plot = FALSE, ann = TRUE, main = "Cluster Dendrogram",sub = NULL, xlab = NULL, ylab = "Height")
```

### 1.

De acuerdo a un análisis preliminar de los datos y el dendograma generado se considera que el número de grupos más adecuado es cuatro formados por: Grupo 1 {1,5,6,7,9}, Grupo 2 {3,8}, Grupo 3 {2}, Grupo 4 {4,10}.

Se toma esta decisión debido a que se considera que en ese punto se maximiza la similitud entre miembros de los grupos así como la disimilitud entre grupos.

Analizando los grupos formados se puede extraer las principales distinciones de estos y caracterizarlos.
- Grupo 1: Se caracteriza por tener una cantidad de peticiones de navegación relativamente baja y una cantidad de búsquedas moderada. Además, este grupo de usuarios agrega generalmente varios productos al carro y todos compran algo. Por último, la cantidad de peticiones de seleccionar es moderada en este grupo. Estos usuarios pordrían ser caracterizados como personas que no tienen muy claro lo que buscan, sin embargo, son decididos a comprar.


- Grupo 2: Este grupo se caracteriza por tener una cantidad relativamente elevadade peticiones de navegación, búsquedas y selecciones. Por otro lado, los clientes agregan pocos productos al carro sin llegar a concretar la compra. 



- Grupo 3: La principal característica de este grupo es que los usuarios realizan pocas navegaciones, búsquedas y selecciones. Además, los usuarios agregan productos al carro y efectúan la compra. Se podría decir que este grupo está compuesto por personas decididas y que saben lo que quieren.


- Grupo 4: Los usuarios de este grupo se caracterizan por realizar muchas navegaciones, búsquedas y selecciones. Los usuarios agregan productos al carro sin llegar a efectuar el pago. Se podría decir que este grupo está constituído por personas indecisas que no buscan nada en particular.

```r
#Análisis de Componentes Principales para graficar los grupos obtenidos
#Calcula plano de componentes principales y grafica las observaciones con el número de su grupo
res <- princomp(datos, cor=TRUE)
print(res)
plot(res)
biplot(res,xlabs=cutree(hd,4))
```

```r
library(rattle)
#grafica los centros de cada grupo en el plano de componentes principales utilizando colores distintos
biplot(res,xlabs=cutree(hd,4))
centros <- centers.hclust(datos,hd,4)
res2 <- predict(res,centros)
print(res2)
points(res2[,1:2],col=1:4,pch=19)
```
